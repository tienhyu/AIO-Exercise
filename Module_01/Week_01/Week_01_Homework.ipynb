```python
#Q1
import math
def calc_f1_score(tp, fp, fn):
  # Check input value is int
    if not isinstance(tp, int) or not isinstance(fp, int) or not isinstance(fn, int):
        if not not isinstance(tp, int):
            print('tp must be int')
        if not not isinstance(fp, int):
            print('fp must be int')
        if not not isinstance(fn, int):
            print('fn must be int')
        return
  # data must be > 0
    if tp >0 and fp > 0 and fn > 0:
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f1_score = 2 * (precision * recall) / (precision + recall)
        print(f'precision is {precision}')
        print(f'recall is {recall}')
        print(f'f1-Score is {f1_score}')
        return f1_score
    else:
        print('tp and fp and fn must be greater than zero')
    return
```

```python
#Q2
import math

# function: sigmoid
def calc_sig(x):
    return 1 / (1 + math.exp(-x))

# function: relu
def calc_relu(x):
    if x <= 0:
        return 0
    else:
        return x

# function: elu
alpha = 0.01
def calc_elu(x):
    if x <= 0:
        return alpha*(math.exp(x) - 1)
    else:
        return x

# Activate function
def calc_activation_func(x, act_name):
  if act_name == "sigmoid":
    return calc_sig(x)
  elif act_name == "relu":
    return calc_relu(x)
  elif act_name == "elu":
    return calc_elu(x)
  else:
    return None

def exercise2():
  x = float(input('Input x = '))
  if not is_number(x): # check number type
    print('x must be a number')
  else:
    act_name = input(('Please input activation function (sigmoid|relu|elu): ')) # get activation function name
    result = calc_activation_func(x, act_name)
    if result is None:
        print(f'{act_name} function is not supportted')
    else:
        print(f'{act_name}: f({x}) = {result}')

exercise2()
```

```python
#Q3
import random
import math

def calc_ae(pred_sample, target_sample):
    return abs(pred_sample - target_sample)

def calc_se(pred_sample, target_sample):
  return (pred_sample - target_sample) ** 2

def exercise3():
    num_samples = input('Input number of samples (integer number) which are generated: ')
    # Check numeric
    if not num_samples.isnumeric():
        print("number of samples must be an integer number")
        return
    loss_name = input('Input loss name (MAE | MSE | RMSE): ')
    loss_name = loss_name.upper() # upper case then accept both lower and upper case

    final_loss = 0
    num_samples = int(num_samples)
    for i in range(num_samples):
        pred_sample = random.uniform(0,10)
        target_sample = random.uniform(0,10)

        if loss_name == 'MAE':
            loss = calc_ae(pred_sample, target_sample)
        elif loss_name == 'MSE' or loss_name == 'RMSE':
            loss = calc_se(pred_sample, target_sample)
        else:
            print(f'{loss_name} regression loss function is not supported')
            return
        final_loss += loss
        print(f'loss_name: {loss_name}, sample: {i}: pred: {pred_sample} target: {target_sample} loss: {loss}')

    final_loss /= num_samples
    if loss_name == 'RMSE':
        final_loss = math.sqrt(final_loss)
    print(f'{loss_name}: {final_loss}')
```

```python
#Q4

import math

def is_number(x):
    try:
        float(x)
    except ValueError:
        return False
    return True

def approx_single_sin(x, n): # single element

    sign = (-1)**n
    numerator = x**(2*n + 1)
    denominator = math.factorial(2*n + 1)
    result = (sign * numerator / denominator)
    return result

def approx_single_cos(x, n): # single element

    sign = (-1)**n
    numerator = x**(2*n)
    denominator = math.factorial(2*n)
    result = (sign * numerator / denominator)
    return result

def approx_single_sinh(x, n): # single element

    sign = 1
    numerator = x**(2*n + 1)
    denominator = math.factorial(2*n + 1)
    result = (sign * numerator / denominator)
    return result

def approx_single_cosh(x, n): # single element

    sign = 1
    numerator = x**(2*n)
    denominator = math.factorial(2*n)
    result = (sign * numerator / denominator)
    return result


def exercise4():
    func_name = input('Input the function that you want to calculate (sin | cos | sinh | cosh): ')
    x = input("Input x in radians:")
    if is_number(x):
        x = float(x)
    else:
        print("x must be a number in radians")
        return

    num_loops = input("Input number of loops (positive integer): ")
    if not num_loops.isnumeric() or int(num_loops) <= 0:
        print("number of samples must be a positive integer")
        return
    result = 0 # Initial value

    num_loops = int(num_loops)
    for i in range(num_loops):

        if func_name == "sin":
            result_element = approx_single_sin(x, i)
        elif func_name == "cos":
            result_element = approx_single_cos(x, i)
        elif func_name == "sinh":
            result_element = approx_single_sinh(x, i)
        elif func_name == "cosh":
            result_element = approx_single_cosh(x, i)
        else:
            print(f'{func_name} function is not supported')
            return
        result += result_element # Accumulate the term value

    print(f'Approximated cos({x}) = {result}')
```

```python
#Q5
import math

def calc_mean_diff_root_nth_error():
    y = float(input("Input y (true) value: "))
    y_hat = float(input("Input y_hat (predict) value: "))
    n = int(input("Input nth rooth: "))
    p = int(input("Input loss function level: "))
    y_root = y**(1/n)
    y_hat_root = y_hat**(1/n)
    difference = y_root - y_hat_root
    loss = difference ** p
    return loss
```
