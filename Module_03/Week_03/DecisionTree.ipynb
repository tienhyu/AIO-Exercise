# **1.0 Decision Tree for Classification**

## **Gini**

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

data = {'Age'           :       [23, 25, 27, 29, 29],
        'Likes English' :       [0, 1, 1, 0, 0],
        'Likes AI'      :       [0, 1, 0, 1, 0],
        'Raise Salary'  :       [0, 0, 1, 1, 0]}

# Load data
df = pd.DataFrame(data)

# Get value and convert to numpy
X = df[['Age', 'Likes English', 'Likes AI']].values
y = df[['Raise Salary']].values.reshape(-1,)

# Define classifier
clf = DecisionTreeClassifier()

# Train
clf = clf.fit(X, y)
```

```python
X
```

```python
y
```

```python
# Test prediction

x_test = np.array([[27, 1, 1]])
predicted_label = clf.predict(x_test)
predicted_label
```

```python
from sklearn.tree import plot_tree

# Visualization
plot_tree(clf, feature_names=['Age', 'Likes English', 'Likes AI'], fontsize=10)
```

## **Entropy-Information Gain**

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

data = {'Age'           :       [23, 25, 27, 29, 29],
        'Likes English' :       [0, 1, 1, 0, 0],
        'Likes AI'      :       [0, 1, 0, 1, 0],
        'Raise Salary'  :       [0, 0, 1, 1, 0]}

# Load data
df = pd.DataFrame(data)

# Get value and convert to numpy
X = df[['Age', 'Likes English', 'Likes AI']].values
y = df[['Raise Salary']].values.reshape(-1,)

# Define classifier
clf = DecisionTreeClassifier(criterion='entropy')

# Train
clf = clf.fit(X, y)
```

## **IRIS Dataset**

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.tree import DecisionTreeClassifier
```

```python
# Load the diabetes dataset
iris_X, iris_y = datasets.load_iris(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_X, iris_y, test_size=0.2, random_state=7
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_pred, y_test)
```

```python
from sklearn.tree import plot_tree

# Visualization
plot_tree(clf, feature_names=["Sepal_Length", "Sepal_Width", "Petal_Length", "Petal_Width"])
```

# **2.0 Decision Tree for Regreesion**

## **Sum of Squared Error**

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor

data = {'Age': [23, 25, 27, 29, 29],
        'Likes English': [0, 1, 1, 0, 0],
        'Likes AI': [0, 1, 0, 1, 0],
        'Salary': [200, 400, 300, 500, 400]}

# Load data
df = pd.DataFrame(data)
```

```python
df
```

```python
# Get value and convert to numpy
X = df[['Age', 'Likes English', 'Likes AI']].values
y = df[['Salary']].values.reshape(-1,)

# Define regressor
reg = DecisionTreeRegressor()

# Train
reg = reg.fit(X, y)
```

```python
# Predict
x_test = np.array([[27, 0, 1]])
predicted_label = reg.predict(x_test)
predicted_label
```

```python
from sklearn.tree import plot_tree

# Visualization
plot_tree(reg, feature_names=['Age', 'Likes English', 'Likes AI'], fontsize=6)
```

```python
# Get value and convert to numpy
X = df[['Likes AI']].values
y = df[['Salary']].values.reshape(-1,)

# Define regressor
reg = DecisionTreeRegressor()

# Train
reg = reg.fit(X, y)

from sklearn.tree import plot_tree

# Visualization
plot_tree(reg, feature_names=['Likes AI'], fontsize=6)
```

## **CPU Machine Dataset**

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.datasets import fetch_openml

machine_cpu = fetch_openml(name='machine_cpu')
```

```python
print(machine_cpu.DESCR)
```

```python
x_data = machine_cpu.data
y_data = machine_cpu.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_data, y_data, test_size=0.2, random_state=7
)
```

```python
# Define regressor
reg = DecisionTreeRegressor(max_depth=1)

# Train
reg = reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
mean_squared_error(y_pred, y_test)
```

```python
from sklearn.tree import plot_tree

# Visualization
plot_tree(reg, feature_names=['MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN', 'CHMAX'])
```
